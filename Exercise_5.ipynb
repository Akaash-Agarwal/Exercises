```python
#1 Import the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#2 and 3: Reading the file and naming it
iris = pd.read_csv('iris.data', sep = ',', header = None)
```

```python
#4 Assigning column names
iris.rename(columns = {0:'sepal_lenght (in cim)', 1: 'sepal_width (in cm)', 2: 'petal_length (in cm)', 3: 'petal_width (in cm)', 4: 'Class'}, inplace = True)
iris
```

```python
#5 Finding any missing values
iris.isnull().sum()
```

```python
#6 Lets set the values of the rows 10 to 29 of the column 'petal_length' to NaN
iris['petal_length (in cm)'].iloc[9:29] = np.nan
iris['petal_length (in cm)'].iloc[9:29]
```

```python
#7 Good, now lets substitute the NaN values to 1.0
iris.replace(np.nan, 1.0, inplace = True)
iris['petal_length (in cm)'].iloc[9:29]
```

```python
#8Now let's delete the column class
iris.drop(columns = ['Class'], inplace = True)
iris
```

```python
#9 Set the first 3 rows as NaN
iris.iloc[0:3,:] = np.nan
iris
```

```python
#10 Delete the rows that have NaN
iris.dropna(inplace = True)
iris
```

```python
#11 Reseting the index
iris.reset_index(inplace = True)
iris
```
